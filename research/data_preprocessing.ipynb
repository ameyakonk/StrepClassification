```python
import pandas as pd
import numpy as np
import cv2
import matplotlib as plt
import os
from pathlib import Path
```

```python
image_dir = 'data/'
data = pd.read_csv('dataset_120.csv')
data = data.drop(columns = ['patient_id'])
data['ImageName'] = image_dir + data['ImageName'] 

image_names = data['ImageName']
image_label = data['label']
symptom_features = data.iloc[:, 2:]

print(image_names)
```

## Resizing the image to a fixed size

```python
## Create output folder for the resized_images
output_folder = r'data/resized_images'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

n_x = 256
n_y = 256
for image in image_names:

    ## read image
    img = cv2.imread(image)
    # print("Original Shape: ",img.shape)

    ## resize image to a fixed dimension
    img_resize = cv2.resize(img, (n_x, n_y))
    # print("New Shape: ",img_resize.shape)

    ## save new resized images to the new folder
    save_path = os.path.join(output_folder, Path(image).name)
    print(save_path)
    cv2.imwrite(save_path, img_resize)

    # ## display original and resized images
    # if img is not None:
    #     fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

    #     ax1.imshow(img)
    #     ax2.imshow(img_resize)

    #     ax1.set_title('Image 1')
    #     ax2.set_title('Image 2')

    #     ax1.axis('off')
    #     ax2.axis('off')

    #     plt.tight_layout()

    #     plt.show()
```
